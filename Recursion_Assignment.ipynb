```python
"""
Given an integer `n`, return *`true` if it is a power of two. Otherwise, return `false`*.

An integer `n` is a power of two, if there exists an integer `x` such that `n == 2x`.

**Example 1:**
Input: n = 1 

Output: true

**Example 2:**
Input: n = 16 

Output: true

**Example 3:**
Input: n = 3
"""
class Solution():
    def pwrof2(self,n):
        if n < 0:
            return False
        elif bin(n).count('1') == 1:
            return True
        else:
            return False
```

```python
"""
Given a positive integer, N. Find the factorial of N. 

**Example 1:**

Input: N = 5 

Output: 120

**Example 2:**

Input: N = 4

Output: 24
"""
class Solution():
    def factorialofN(self,n):
        if n == 0:
            return 1
        else:
            prdt = 1
            for i in range(n,0,-1):
                prdt = prdt*i
            return prdt
        
    """ OR """
def factorialofN(n):
    if n==0:
        return 1
    else:
        return n* factorialofN(n-1)
    
#Driver Code
n = 5
factorialofN(n)
```

```python
"""
Given an array of integers **arr**, the task is to find maximum element of that array using recursion.

**Example 1:**

Input: arr = {1, 4, 3, -5, -4, 8, 6};
Output: 8

**Example 2:**

Input: arr = {1, 4, 45, 6, 10, -8};
Output: 45
"""
  
def findMaxRec(A, n):
    if (n == 1):
        return A[0]
    return max(A[n - 1], findMaxRec(A, n - 1))
 
        
A = [1, 4, 45, 6, -50, 10, 2]
n = len(A)
print(findMaxRec(A, n))
```

```python
"""
Given a string S, the task is to write a program to print all permutations of a given string.

**Example 1:**

***Input:***

*S = “ABC”*

***Output:***

*“ABC”, “ACB”, “BAC”, “BCA”, “CBA”, “CAB”*

**Example 2:**

***Input:***

*S = “XY”*

***Output:***

*“XY”, “YX”*
"""


def toString(a):
    return ''.join(a)

def permute(a, l, r):
    if l == r:
        print(toString(a))
    else:
        for i in range(l, r):
            tmp = a[i]
            a[i] = a[l]
            a[l] = tmp
            
            permute(a, l+1, r)
            tmp = a[i]
            a[i] = a[l]
            a[l] = tmp  # backtrack
 
 
# Driver code
string = "ABC"
l = 0
r = len(string)
a = list(string)
 

permute(a, l, r)
```

```python
"""
Given an array, find a product of all array elements.

**Example 1:**

Input  : arr[] = {1, 2, 3, 4, 5}
Output : 120
**Example 2:**

Input  : arr[] = {1, 6, 3}
Output : 18
"""

def prod_of_array(lst,n):
    if n == 0:
        return 1
    else:
        return lst[n]*prod_of_array(lst,n-1)

#Driver Code

lst = [1, 2, 3, 4, 5]
n = len(lst) - 1
print(prod_of_array(lst,n))
```
