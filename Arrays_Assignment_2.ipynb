```python
"""
**Question 1**
Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2),..., (an, bn) 
such that the sum of min(ai, bi) for all i is maximized. 
Return the maximized sum.

**Example 1:**
Input: nums = [1,4,3,2]
Output: 4

**Explanation:** All possible pairings (ignoring the ordering of elements) are:

1. (1, 4), (2, 3) -> min(1, 4) + min(2, 3) = 1 + 2 = 3
2. (1, 3), (2, 4) -> min(1, 3) + min(2, 4) = 1 + 2 = 3
3. (1, 2), (3, 4) -> min(1, 2) + min(3, 4) = 1 + 3 = 4
So the maximum possible sum is 4
"""

class Solution(object):
    def groupbysum(self, nums):
        nums.sort()             #this will gives all the numbers we can pick without grouping them. 
                                #Because the min of every number would be the odd number.
        # As observed in the above example, we can see that max sum is acheived with the sorted group only.
        
        return sum(nums[::2])   # This helps us select those numbers.
```

```python
"""
Question 2
Alice has n candies, where the ith candy is of type candyType[i]. 
Alice noticed that she started to gain weight, so she visited a doctor. 

The doctor advised Alice to only eat n / 2 of the candies she has (n is always even).
Alice likes her candies very much, and she wants to eat the maximum number of different types of candies 
while still following the doctor's advice. 

Given the integer array candyType of length n, return the maximum number of different types of candies she can eat 
if she only eats n / 2 of them.

Example 1:
Input: candyType = [1,1,2,2,3,3]
Output: 3

Explanation: Alice can only eat 6 / 2 = 3 candies. Since there are only 3 types, she can eat one of each type.
"""

class Solution(object):
    def Candies(self, candyType):
        count = 1
        max_allowed = int(len(candyType)/2)
        candyType.sort()
        for i in range(1,len(candyType)):
            if candyType[i] != candyType[i-1]:  # This is to check how many unique candies are there in the list
                count += 1

            if count == max_allowed:          # We need to apply this because, if the max is met, there is no use to calculate rest as that is max
                break
        return min(count,max_allowed)  #we need to apply this to satisfy this condition where the list has only two items and both are unique
                                        #in that case, we get count 2 where as max allowed would be 1.
```

```python
"""
Question 3
We define a harmonious array as an array where the difference between its maximum value
and its minimum value is exactly 1.

Given an integer array nums, return the length of its longest harmonious subsequence
among all its possible subsequences.

A subsequence of an array is a sequence that can be derived from the array by deleting some or no elements 
without changing the order of the remaining elements.

Example 1:
Input: nums = [1,3,2,2,5,2,3,7]
Output: 5

Explanation: The longest harmonious subsequence is [3,2,2,2,3].
"""
from collections import Counter
def Solution(object):
    def longest_harmonius(nums):
        longest_list_count = 0
        h = Counter(nums)       #Used to create a dictionary to and count the instances, we get the counter library from collections
        
        for key,val in h.items():
            if key+1 in h:      # Since harmonious means max - min == 1, we are checking if we have keys which satisfy that condition
            
                longest_list_count = max(longest_list_count,(val + h[key+1]))
            
        return longest_list_count
```

```python
"""
Question 4
You have a long flowerbed in which some of the plots are planted, and some are not.
However, flowers cannot be planted in adjacent plots.
Given an integer array flowerbed containing 0's and 1's, where 0 means empty and 1 means not empty,and an integer n, 
return true if n new flowers can be planted in the flowerbed without violating the no-adjacent-flowers rule and false otherwise.

Example 1:
Input: flowerbed = [1,0,0,0,1], n = 1
Output: true
"""
class Solution(object):
    def canPlaceFlowers(self,nums,n):
        number_of_flowers_can_plot = 0
        length = len(nums)
        for i in range(length):
            if nums[i] == 0:
                left = (i==0) or (nums[i - 1] == 0)
                right =  (i == length -1) or (nums[i+1] == 0) #If we write this condition in the reverse order, we get index out of range exception.
                if left and right:
                    nums[i] = 1                      # If we dont write this, it will again work as true in the left and right conditions.
                    number_of_flowers_can_plot += 1
        
        return number_of_flowers_can_plot >= n
```

```python
"""
Question 5
Given an integer array nums, find three numbers whose product is maximum and return the maximum product.

Example 1:
Input: nums = [1,2,3]
Output: 6
"""
class Solution(object):
    def maximumProduct(self, nums):
        nums.sort()
        length = len(nums)    
        max_product_1 = (nums[length - 3])*(nums[length - 2]) * (nums[length-1])
        max_product_2 = (nums[length-1])*(nums[0])*(nums[1]) 
        return max(max_product_1,max_product_2)
```

```python
"""
Question 7
An array is monotonic if it is either monotone increasing or monotone decreasing.

An array nums is monotone increasing if for all i <= j, nums[i] <= nums[j]. An array nums is
monotone decreasing if for all i <= j, nums[i] >= nums[j].

Given an integer array nums, return true if the given array is monotonic, or false otherwise.

Example 1:
Input: nums = [1,2,2,3]
Output: true
"""
class Solution(object):
    def isMonotonic(self, nums):
        lst = sorted(nums)
        if lst == nums or lst == nums[::-1]:
            return True
        else:
            return False
```

```python
"""
Question 8
You are given an integer array nums and an integer k.

In one operation, you can choose any index i where 0 <= i < nums.length and 
change nums[i] to nums[i] + x where x is an integer from the range [-k, k]. 
You can apply this operation at most once for each index i.

The score of nums is the difference between the maximum and minimum elements in nums.

Return the minimum score of nums after applying the mentioned operation at most once for each index in it.

Example 1:
Input: nums = [1], k = 0
Output: 0

Explanation: The score is max(nums) - min(nums) = 1 - 1 = 0.
"""
class Solution:
    def smallestRangeI(self, nums, k):
        if len(nums) == 1:
            return 0

        if (max(nums) - k) - (min(nums) + k ) <= 0:
            return 0
        else:
            return (max(nums) - k) - (min(nums) + k )
```
